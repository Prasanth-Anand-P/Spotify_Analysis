```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tr = pd.read_csv('tracks.csv')
df_tr.head()
```

```python
#check for null
pd.isnull(df_tr).sum()
```

```python
#Getting information about the data frame
df_tr.info()
```

```python
#checking the 10 least popular song
sort_tr = df_tr.sort_values("popularity", ascending = True).head(10)
sort_tr
```

```python
df_tr.describe().transpose()
```

```python
most_popular = df_tr.query('popularity>90', inplace=False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tr.set_index("release_date", inplace=True)
df_tr.index=pd.to_datetime(df_tr.index)
df_tr.head()
```

```python
df_tr[['artists']].iloc[18]
```

```python
df_tr['durations'] = df_tr['duration_ms'].apply(lambda x: round(x/1000))
df_tr.drop('duration_ms', inplace=True, axis = 1)
```

```python
df_tr.durations.head()
```

```python
corr_df = df_tr.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='black')
heatmap.set_title("Correlation Heatmap between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=300)
```

```python
sample_df = df_tr.sample(int(0.004*len(df_tr)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x="energy", y="loudness", color='red').set_title("Loudness VS Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x="popularity", y="danceability", color='red').set_title("Popularity VS Danceability")
```

```python
df_tr['dates'] = df_tr.index.get_level_values('release_date')
df_tr.dates = pd.to_datetime(df_tr.dates)
years = df_tr.dates.dt.year
```

```python
sns.displot(years, discrete = True, aspect=2, height=5, kind = 'hist').set(title = "No. of songs per year")
```

```python
total_dr = df_tr.durations
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year VS Duration")
plt.xticks(rotation = 90)
```

```python
total_dr = df_tr.durations
sns.set_style(style = "whitegrid")
fig_dims = (10,5)
fig,ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title = "Year VS Duration")
plt.xticks(rotation = 60)
```

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
```

```python
plt.title("Duration of the Songs in different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = "genre", x = "duration_ms", data = df_genre)
plt.ylabel("Duration in MS")
plt.ylabel("Genres")
```
